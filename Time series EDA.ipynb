## Time Resampling

```python
df_tesla.head()
```

```python
# Rule A: year and frequency
df_tesla.resample(rule='A')
```

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
# quaterly start frequency 
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
df_tesla.resample(rule='BA').max()
```

```python
# Busness end frequency 
df_tesla.resample(rule='BA').max()['High'].plot()
```

```python
df_tesla.resample(rule='BQS').max()
```

```python
## Ploting
df_tesla['Open'].resample(rule='BA').max().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').mean()
```

```python
df_tesla['Open'].resample(rule='M').mean().plot(kind='bar',figsize=(20,9))
```

```python
 df_tesla['High'].rolling(10).mean().head(20)
```

```python
df_tesla['Open 30 days rolling'] = df_tesla['Open'].rolling(30).mean()
df_tesla.head(32)
```

```python
df_tesla[['Open','Open 30 days rolling']].plot(figsize=(20,10))
```
