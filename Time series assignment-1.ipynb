### Assignment:
1. Read the microsoft data using Pandas Data Reader.
2. Get the maximum price of the share from 2017 to 2022.
3. Which is the day of highest price of the stock?
4. Which is the day of lowest price of the stock?

```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start_time = datetime.now()
```

```python
df_msft = pdr.get_data_yahoo('MSFT')
df_msft.head()
```

```python
df_msft.plot(figsize=(20,10))
```

```python
df_msft[['High','Low']].plot(ylim=[50,350], figsize=(20,10), ls='--')
```

```python
df1 = df_msft.loc['2017-07-10':'2022-07-08']
print('*Maximum price of the share from 2017 to 2022: ',df1['High'].max())
```

```python
df2 = df_msft[df_msft['High']>=349]
df2 = df2.reset_index()
df2
```

```python
print('*The day of highest price of the stock: ',df2.iat[0,0])
```

```python
print('Minimum price of the share from 2017 to 2022: ',df1['Low'].min())
```

```python
df3 = df_msft[df_msft['Low']<=69.2]
df3 = df3.reset_index()
df3
```

```python
print('*The day of lowest price of the stock: ',df3.iat[0,0])
```

```python
end_time = datetime.now()
```
